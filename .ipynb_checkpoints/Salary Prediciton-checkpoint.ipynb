```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Data Analyst.xls')
data['Salary']=data['Salary'].str.replace('  Not disclosed ','0')
```

```python
data.describe()
```

```python
for i in range(len(data['Salary'])): 
    if data['Salary'][i]=='0':
        data['Salary'][i]=0
    else:
        data['Salary'][i]=int(''.join((str(data['Salary'][i]).split('-'))[1].split(' ')[1].split(',')))              
```

```python
data['Location']=data['Location'].str.extract('(\w+)')
```

```python
mean = int(np.mean(data['Salary']))
data['Salary']= data['Salary'].replace(0,mean)
```

```python
data
```

```python
for i in range(len(data['Skills'])):
   data['Skills'][i]=data['Skills'][i].split(',')
```

```python
data
```

```python
for i in range(len(data['Experience'])):
    temp=data['Experience'][i].split('-')
    temp[1]=temp[1].split(' ')[0]
    data['Experience'][i] = temp
```

```python
data
```

```python
data.plot(kind='line' , sharex=False , sharey=False)
```

```python
sns.pairplot(data, hue='Location',size=15)
```

```python
for i in range(len(data['Experience'])):
    data['Experience'][i]=int(int(data['Experience'][i][0])+int(data['Experience'][i][0])/2)
```

```python
sns.pairplot(data, hue='Experience',size=10)
```
